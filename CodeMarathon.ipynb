```python
import requests
import json

def iso_duration_to_seconds(iso_duration):
    total_seconds = 0
    if iso_duration.startswith('PT'):
        iso_duration = iso_duration[2:]

    hours, minutes, seconds = 0, 0, 0
    if 'H' in iso_duration:
        hours, iso_duration = iso_duration.split('H', 1)
        total_seconds += int(hours) * 3600
    if 'M' in iso_duration:
        minutes, iso_duration = iso_duration.split('M', 1)
        total_seconds += int(minutes) * 60
    if 'S' in iso_duration:
        seconds, iso_duration = iso_duration.split('S', 1)
        total_seconds += int(seconds)
    
    return total_seconds

def convert_seconds_to_hms(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return f"{hours}h {minutes}m {seconds}s"

x = 0
lim = 100
res = []
max_ = 500  # Limite pour l'exemple

while x <= max_:
    url = f"https://resultscui.active.com/api/results/events/RuninLyonbyHarmonieMutuelle2024/participants?groupId=1045199&routeId=178601&offset=0&limit=100"
    response = requests.get(url)

    if response.status_code == 200:
        json_data = response.json()
        extracted_data = []

        for item in json_data['items']:
            first_name = item['person']['firstName']
            lastName = item['person']['lastName']
            gender = item['person']['gender']
            age = item['person']['age']
            final_result_seconds = iso_duration_to_seconds(item['finalResult']['finalResult'])
            final_result_hms = convert_seconds_to_hms(final_result_seconds)

            data_point = {
                'firstName': first_name,
                'lastName' : lastName,
                'gender': gender,
                'age': age,
                'finalResult': final_result_hms
            }

            extracted_data.append(data_point)

        res.append(json.dumps(extracted_data, indent=4))
        with open('top100.json', 'w') as json_file:
            json.dump(extracted_data, json_file, indent=4)
        x += lim
    else:
        print(f"Failed to retrieve data: {response.status_code} - {response.text}")
        break  # Arrter la boucle si l'API renvoie une erreur 404
print(res)
```

# Connexion  snowflake

```python
import json
import pandas as pd
import snowflake.connector

# Charger les donnes JSON dans un DataFrame Pandas
with open('extracted_data.json', 'r') as json_file:
    data = json.load(json_file)

df = pd.DataFrame(data)

# Connexion  Snowflake
conn = snowflake.connector.connect(
    user='<YOUR_USERNAME>',
    password='<YOUR_PASSWORD>',
    account='<YOUR_ACCOUNT>',
    warehouse='<YOUR_WAREHOUSE>',
    database='<YOUR_DATABASE>',
    schema='<YOUR_SCHEMA>'
)

# Cration d'un curseur pour excuter des commandes SQL
cursor = conn.cursor()

# (Facultatif) Cration de la table dans Snowflake si elle n'existe pas dj
cursor.execute('''
CREATE TABLE IF NOT EXISTS TABLE_RESULTAT (
    firstName STRING,
    gender STRING,
    age INTEGER,
    finalResult STRING
)
''')

# Transfert des donnes du DataFrame vers Snowflake
for index, row in df.iterrows():
    cursor.execute('''
        INSERT INTO TABLE_RESULTAT (firstName, gender, age, finalResult)
        VALUES (%s, %s, %s, %s)
    ''', (row['firstName'], row['gender'], row['age'], row['finalResult']))

# Vrification de l'insertion des donnes
cursor.execute("SELECT * FROM TABLE_RESULTAT LIMIT 5")
print(cursor.fetchall())

# Fermer le curseur et la connexion
cursor.close()
conn.close()
```
